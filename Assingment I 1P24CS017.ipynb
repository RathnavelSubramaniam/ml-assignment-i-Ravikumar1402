(b) What is the range of each quantitative predictor? You can answer this using the range() function. 

```python
column = ['cylinders','weight','mpg','displacement','acceleration','year','horsepower']
min = Auto[column].min()
max = Auto[column].max()
print(f"Range of {column}: {max}, {min}")
```

(c) What is the mean and standard deviation of each quantitative predictor? 

```python
Mean = Auto[column].mean()
Stdev = Auto[column].std()
print(f"Mean of: {column}: {Mean} , Std Dev of: {column}: {Stdev}")
```

(d)	 Now remove the 10th through 85th observations. What is the range, mean, and standard deviation of each predictor in the subset of the data that remains? 

```python
data = Auto.drop(Auto.index[9:85])

Mean = data[column].mean()
Stdev = data[column].std()
Max = data[column].max()
Min = data[column].min()
print(f"Range of {column}: {Max}, {Min}")
print(f"Mean of {column}: {Mean}, Std Dev of {column}: {Stdev}")
```

(e)	Using the full data set, investigate the predictors graphically, using scatterplots or other tools of your choice. Create some plots highlighting the relationships among the predictors. Comment on your findings.

```python
sns.boxplot(x=Auto['mpg'], y=Auto['cylinders'])
plt.title("Boxplot of MPG by cylinders")
plt.show()
```

(f) Suppose that we wish to predict gas mileage (mpg) on the basis of the other variables. Do your plots suggest that any of the other variables might be useful in predicting mpg? Justify your answer

```python
sns.lineplot(x='mpg', y='displacement', data=Auto)
plt.title('MPG vs Displacement - Line Plot')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Displacement')
plt.show()
```

```python
sns.lineplot(x='mpg', y='weight', data=Auto)
plt.title('MPG vs Weight - Line Plot')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Weight')
plt.show()
```

```python
sns.lineplot(x='mpg', y='acceleration', data=Auto)
plt.title('MPG vs Acceleration - Line Plot')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Acceleration')
plt.show()
```

```python
sns.lineplot(x='mpg', y='year', data=Auto)
plt.title('MPG vs Year - Line Plot')
plt.xlabel('Miles Per Gallon')
plt.ylabel('Year')
plt.show()
```

The plots suggest that weight, horsepower, displacement, and cylinders are strong predictors of mpg. Model year and origin may also contribute moderately. Acceleration seems less useful based on visual inspection.
